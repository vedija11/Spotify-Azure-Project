#### **AUTOLOADER**

```python
df_user = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format", "parquet")\
            .option("cloudFiles.schemaLocation", "abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimUser/checkpoint")\
            .option("schemaEvolutionMode", "addNewColumns")\
            .load("abfss://bronze@vedstorageazureproject.dfs.core.windows.net/DimUser")
```

```python
df_user = df_user.withColumn("user_name", upper(col("user_name")))
# display(df_user)
```

```python
df_user_obj = reusable()
df_user = df_user_obj.dropColumns(df_user, ['_rescued_data'])
df_user = df_user.dropDuplicates(['user_id'])
```

```python
# display(df_user, checkpointLocation="abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimUser/checkpoint")
```

```python
df_user.writeStream.format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", "abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimUser/checkpoint") \
    .trigger(once=True) \
    .option("path", "abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimUser/data") \
    .toTable("spotify_cata.silver.DimUser")
```

## **DimTrack**

```python
df_track = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format", "parquet")\
            .option("cloudFiles.schemaLocation", "abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimTrack/checkpoint")\
            .option("schemaEvolutionMode", "addNewColumns")\
            .load("abfss://bronze@vedstorageazureproject.dfs.core.windows.net/DimTrack")
```

```python
df_track = df_track.withColumn("durationFlag", when(col('duration_sec') < 150, "low")\
                                                .when(col('duration_sec') < 300, "medium")\
                                                .otherwise("high"))
df_track = df_track.withColumn("track_name", regexp_replace(col('track_name'), '-', ' '))
df_track = reusable().dropColumns(df_track, ['_rescued_data'])
```

```python
# display(df_track, checkpointLocation="abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimTrack/checkpoint")
```

```python
df_track.writeStream.format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", "abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimTrack/checkpoint") \
    .trigger(once=True) \
    .option("path", "abfss://silver@vedstorageazureproject.dfs.core.windows.net/DimTrack/data") \
    .toTable("spotify_cata.silver.DimTrack")
```
